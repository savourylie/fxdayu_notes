```python
from datetime import datetime
```

## DataAPI.candle()
## 股票K线数据

### 获取一只股票("000001.XSHE")在一段时间(start, end)的日线(freq="D")数据

```python
DataAPI.candle("000001.XSHE", freq='D', start=datetime(2016, 1, 1), end=datetime(2016, 1, 31))
```

### 获取指定长度的数据(length)

```python
DataAPI.candle("000001.XSHE", 'D', start=datetime(2016, 1, 1), length=10)
```

### 获取指定字段(fields)的数据

```python
DataAPI.candle("000001.XSHE", 'D', fields=("close", "volume"), start=datetime(2016, 1, 1), length=10)
```

### 获取多只股票的数据

```python
DataAPI.candle(("000001.XSHE", "600000.XSHG"), 'D', start=datetime(2016, 1, 1), length=10)
```

## DataAPI.factor()
## 获取因子数据(日线级别)

```python
fields = ('PE', "PB", "ROE")
DataAPI.factor("000001.XSHE", fields, start=datetime(2017, 5, 1), end=datetime(2017, 5, 15, 15))
```

```python
DataAPI.factor(("000001.XSHE", "000002.XSHE"), fields=fields, start=datetime(2017, 5, 1), end=datetime(2017, 5, 15, 15))
```
