````python
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fxdayu_data import DataAPI
from datetime import datetime
import talib as ta

%matplotlib inline
````

# TA_Lib
1. 导入 DataAPI
2. 读取 600036 股票的 open/high/close/low/volume
3. 用 talib 计算 EMA20，EMA60 并用图将 close 和 EMA 显示出来

````python
start_date = datetime(2014, 1, 1)
end_date = datetime(2017, 1, 1)

prices = DataAPI.candle("600036.XSHG", "D", start=start_date, end=end_date)
````

````python
type(prices)
````

````python
prices.head()
````

````python
prices['EMA20'] = ta.abstract.EMA(prices, 20)
prices['EMA60'] = ta.abstract.EMA(prices, 60)
````

````python
prices.head(30)
````

````python
fig = plt.figure(figsize=(16, 9))
plt.plot(prices['close'], c='black')
plt.plot(prices['EMA20'], alpha=0.5)
plt.plot(prices['EMA60'], alpha=0.5)
plt.legend()
plt.show()
````

# Series 合成 DataFrame

1. 读取 3 只股票的 close
2. 计算 3 只股票的 EMA20
3. 将 3 只股票的 EMA20 Series 合成 DataFrame

````python
stocks_prices = DataAPI.candle("HS300", "D", start=start_date, end=end_date)
````

````python
df_cols = stocks_prices.items[[2, 3, 5]]
df_close = stocks_prices.minor_xs('close')[df_cols]
````

````python
df_close = df_close.fillna(method='ffill')
````

````python
for col in df_close.columns:
    df_close[col + '_EMA'] = ta.EMA(df_close[col].values, 20)
````

````python
df_ema = df_close.iloc[:, 3:]
````
